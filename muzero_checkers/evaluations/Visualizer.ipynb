```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
eval0 = np.load('evaluation_against_random_0.npy')
eval1 = np.load('evaluation_against_random_1.npy')

print(eval0, eval1)
```

```python
wins_as_starting_player0 = sum(eval0[0,:])
wins_as_second_player0 = sum(eval0[2,:])

wins_as_starting_player1 = sum(eval1[0,:])
wins_as_second_player1 = sum(eval1[2,:])

plt.plot(np.arange(2), [wins_as_second_player0/50, wins_as_second_player1/50], label="Win percentage as starting player")
plt.plot(np.arange(2), [wins_as_starting_player0/50, wins_as_starting_player1/50], label="Win percentage as second player")
plt.legend()
plt.xlabel("Evaluation Number")
plt.ylabel("Win percentage (50 games)")
plt.title("Evaluation Results")
plt.show()
```
